# DATA IMPUTATION AND MANIPULATION

```python
#define function of impute_median
def impute_median(series):
    return series.fillna(series.median())
```

```python
df.Rating = df['Rating'].transform(impute_median)
```

```python
df.isnull().sum()
```

```python
df['Content Rating']

print(df['Content Rating'].mode())
df['Content Rating'].fillna(str(df['Content Rating'].mode().values[0]), inplace=True)
```

```python
df.isnull().sum()
```

```python
#modes of categorical values

print(df['Type'].mode())

print(df['Current Ver'].mode())
print(df['Android Ver'].mode())
```

```python
df['Type'].fillna(str(df['Type'].mode().values[0]), inplace=True)
```

```python
df['Current Ver'].fillna(str(df['Current Ver'].mode().values[0]), inplace=True)
df['Android Ver'].fillna(str(df['Android Ver'].mode().values[0]), inplace=True)
```

```python
df.isna()
```

```python
df.isnull().sum()
```

```python
df.head(10)
```

```python
df['Price'] = df['Price'].apply(lambda x: str(x).replace('$', ' ') if '$' in str(x) else str(x))
```

```python
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

```python
df['Price']
```

```python
df['Price'] = df['Price'].apply(lambda x: float(x))
```

```python
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
```

```python
df.info()
```

```python
df['Installs'] = df['Installs'].apply(lambda x: str(x).replace('+', ' ') if '+' in str(x) else str(x))
```

```python
df['Installs']
```

# DATA VISUALIZATION

```python
grp = df.groupby('Category')

x = grp['Rating'].agg(np.mean)
y = grp['Price'].agg(np.sum)
z = grp['Reviews'].agg(np.mean)
print(x)
print(y)
print(z)
```

```python
plt.figure(figsize=(16, 7))
plt.plot(z, 'r--', color='green')
plt.plot(x)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(16, 7))
plt.plot(z, 'ro', color='green')
plt.plot(x)
plt.xticks(rotation=90)
plt.xlabel('Categories')
plt.ylabel('Reviews')
plt.title('Category wise reviews')
plt.show()
```

```python
plt.figure(figsize=(18, 6))
plt.plot(z, )
plt.plot(y)
plt.xticks(rotation=90)
plt.xlabel('Categories')
plt.ylabel('Price')
plt.title('Price value category wise')
plt.show()
```

```python
plt.figure(figsize=(18, 8))
plt.plot(z, 'r--', color='purple')
plt.xticks(rotation=90)
plt.xlabel('Categories--->')
plt.ylabel('Reviews--->')
plt.title('Review for Categories')
plt.show()
```
